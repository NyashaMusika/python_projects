```python
#note for averages use average function during generation
class Conversion_cycle():

    def __init__(self,c_assets,c_liabilities,inventory,sales,average_recivables,COGs,average_payables,average_inventory):
        self.c_assets = c_assets
        self.c_liabilities = c_liabilities
        self.inventory = inventory
        self.sales = sales
        self.average_inventory = average_inventory
        self.average_payables = average_payables
        self.average_recivables = average_recivables
        self.COGs = COGs

    def current_ratio(self):
        return self.c_assets/self.c_liabilities
    def quick_ratio(self):
        return (self.c_assets-self.inventory)/self.c_liabilities
    def recievables(self):
        return self.sales/self.average_recivables
    def payables(self):
        return self.COGs/self.average_payables
    def inventory_turnover(self):
        return self.COGs/self.average_inventory
    def conversion_cycle(self):
        return (365/self.recievables()) + (365/self.inventory_turnover()) - (365/self.payables())
```

```python
def averages(a,b):
    return (a+b)/2
```

```python
#note for averages use average function during generation
class Asset_equity_turnover():
    
    def __init__(self,sales,average_t_assets,average_f_assets,average_equity):
        self.sales = sales
        self.average_t_assets = average_t_assets
        self.average_f_assets = average_f_assets
        self.average_equity = average_equity
        
    def total_asset_turnover(self):
        return self.sales/self.average_t_assets
    def fixed_asset_turnover(self):
        return self.sales/self.average_f_assets
    def equity_turnover(self):
        return self.sales/self.average_equity
        
        
```

```python
class Profit_margin():
    
    def __init__(self,gross_profit,sales,income,operating_profit,interest,average_total_capital,average_equity,pref_div):
        self.gross_profit = gross_profit
        self.sales = sales
        self.income = income 
        self.operating_profit = operating_profit
        self.interest = interest
        self.average_total_capital = average_total_capital
        self.average_equity = average_equity
        self.pref_div = pref_div
        
    def gross_profit_margin(self):
        return self.gross_profit/self.sales
    def net_profit_margin(self):
        return self.income/self.sales
    def operating_profit_margin(self):
        return self.operating_profit/self.sales
    def return_on_total_capital(self):
        return (self.income-self.interest)/self.average_equity
```

```python
class Debt():
    
    def __init__(self,total_LT_debt,total_equity,total_ST_debt,total_assets):
        self.total_LT_debt = total_LT_debt
        self.total_equity = total_equity
        self.total_ST_debt = total_ST_debt
        self.total_assets = total_assets
        
    def debt_to_equity(self):
        return self.total_LT_debt/self.total_equity
    def debt_ratio(self):
        return (self.total_LT_debt+self.total_ST_debt)/self.total_assets
    
```

```python
def cash_flow(cash_flow,total_LT_debt):
    return cash_flow/total_LT_debt
```
